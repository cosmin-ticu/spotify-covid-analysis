## How have viral Spotify songs' charactheristics changed throughout the COVID pandemic while cases were rising?

#### Has there been a decline in danceability or valence?

By looking at the below line chart we can say that danceability declined marginally meanwhile valence increased marginally. Altogether, we could say that the value for these audio features were fluctuating over the year.

#### Can we identify certain seasonalities or spikes within the data that match COVID-19 noteable moments? (Introduction of quarantine restrictions)

The below three audio features were deemed the most important by us when answering this question. Please see the definition of these audio features:

- valence - A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry). 

- danceability - Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable. 

- energy - Energy is a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy. For example, death metal has high energy, while a Bach prelude scores low on the scale. Perceptual features contributing to this attribute include dynamic range, perceived loudness, timbre, onset rate, and general entropy.

```python
# Importing ther necessary libraries
import matplotlib.pyplot as plt # this is used for the plot the graph 
import seaborn as sns # used for plot interactive graph.
%matplotlib inline
```

```python
# Extracting weeknumber from Date for easier visualization
import datetime
df["Date"] = pd.to_datetime(df["Date"],
                           errors = 'coerce')
df["week"] = df["Date"].dt.isocalendar().week
```

We can see that valence, energy and danceability are all fluctuating throughout the whole year but we can identify a huge drop in both features around week 13-15 when most of the countries imposed travel restrictions. 

Other than that it could be said that valence, energy and danceability are moving together but there are some irregularities in this as well. The best example is week 37 when we can identtify a huge drop in danceability and a huge spike in valence and energy. Although, I could not find any important event that could explain this irregularity.  

```python
# Line chart for certain audio features over the period
import numpy as np
plt.figure(figsize=(16, 8))
sns.set(style="whitegrid")
columns = ["danceability", "valence", "energy"]
for col in columns:
    x = df.groupby("week")[col].mean() #Groupping by week for more digestable chart
    ax= sns.lineplot(x=x.index,y=x,label=col)
ax.set_title('Audio features over the period')
ax.set_ylabel('Value')
ax.set_xlabel('Week number')
plt.xticks(np.arange(1, len(x)+1, 2)); #Changed the ticks frequency for easier interpretation
```

```python
# Dual axis just for fun and showcasing our plotting abilities
plt.figure(figsize=(40, 20))
fig, ax1 = plt.subplots()

color = 'tab:orange' # first y axis
ax1.set_xlabel('week')
ax1.set_ylabel('danceability', color=color)
ax1.plot(df.groupby("week")['danceability'].mean(), color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:purple'
ax2.set_ylabel('valence', color=color)  # we already handled the x-label with ax1
ax2.plot(df.groupby("week")['valence'].mean(), color=color)
ax2.tick_params(axis='y', labelcolor=color)
```

#### Artists that benefitted most from COVID

We can identify several things regarding the artists performance for the year 2020. By top performing artists we meant those who featured in Viral Top 50 the most times, it can be through one song or several songs.

```python
# Grouping by artists and showing the unique tracks produced, number of times featured in the charts, mean danceability and mean valence
top_artists = df.groupby(
   ['Artist']
).agg(
    { 
        'Track Name':'nunique',
        'trackID':'count', 
        'danceability':'mean',  
         'valence':'mean'
    }
)
```

The table below shows us the top 15 performing artists. We should higlight that we have two columns regarding features. 

- The column named 'Unique tracks' means that how many different tracks of the given artist were featured in the Viral Top 50.

- The column named 'Number of times featured' means the total number of features in the Viral Top 50 charts. In this case it can happen that a song featured not only in one week but in several weeks.

The most noticable thing is that an artist is dominating both charts. RBD had 27(!!) different songs featured for a total of 31 times throughout 2020. CORPSE is following RBD with a total number of 29 features but they managed that many features with only 6 songs. This means that CORPSE's songs were more durable over the time, meanwhile RBD's songs were basically feaured only once.

```python
# Artists ordered by number of times featured and the top 15 shown
top_artists.sort_values("Number of times featured", ascending = False).head(15)
```

#### Interactive chart by using bokeh

For creating an interactive chart I used the library called bokeh and one of the charts above. In this chart if you hover over the bars you could see other features for each artist, like average danceability, average valence and best position.

- average danceability is calculated by taking all the danceability scores of each song of the artist (some artists have more songs) and taking the average
- average valence is calculated by taking all the valence scores of each song of the artist (some artists have more songs) and taking the average
- best position is calculated by looking for the best position of any song of the artist (again some artists have more songs)

```python
import bokeh
from bokeh.io import show, output_notebook
from bokeh.plotting import figure, ColumnDataSource
from bokeh.models import HoverTool
output_notebook()

toplot  = df.groupby(
   ['Artist']
).agg(
    {  
        'trackID':'count', 
        'danceability':'mean',  
         'valence':'mean',
        'Position':'min'
    }
).sort_values("trackID", ascending = False).head(15) 

source = ColumnDataSource(toplot)
artists = source.data['Artist'].tolist()
p = figure(x_range = artists)

p.title.text = 'Artists per number of times featured - interactive chart'

p.vbar(x= 'Artist',
       top = 'trackID',
       source = source,
       color='orange',
       width = 0.5)

p.xaxis.major_label_orientation = "vertical"

hover = HoverTool(tooltips=[("average danceability", "@danceability"),
              ("average valence", "@valence"),              
              ("best position", "@Position")])
p.add_tools(hover)

show(p)

```

#### Which charachteristic saw the largest change? (Most fluctuation in features)

```python
# Getting the average audio features value for the top 5 songs
weekly_top5_chars = df[df['Position'] < 6]
weekly_top5_chars = weekly_top5_chars.groupby(['Date']).agg(
    {
         'danceability':'mean',
         'energy':'mean',
         'key':'mean',
         'loudness':'mean',
         'acousticness':'mean',
         'instrumentalness':'mean',
         'liveness':'mean',
         'valence':'mean',
         'tempo':'mean',
         'speechiness':'mean'
    }
)
```

```python
# Defining measures for a for loop used later
measures = ['danceability', 'energy', 'key',
       'loudness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'speechiness']
```

```python
# Plotting all the audio features on a different line chart by using a for loop
for measure in measures:
    weekly_top5_chars = df.groupby(['Date']).mean(measure).reset_index()
    plt.figure(figsize=(12,5))
    sns.lineplot(data=weekly_top5_chars, x='Date',y=measure)
    plt.title(measure)
```

## What have been the trending charachteristics of songs with sustained popularity?

#### How has duration changed over the past year?

Duration seemed to fluctuate throughout 2020 between 2.9 and 3.5 minutes. We can see that people listenned to mainly shorter songs in November followed by longer songs in December. Maybe the Christmas songs hav longer duration compared to any other genre of songs. 

```python
# Group the data frame by week and extract a number of stats from each group
weekly_stats = df.groupby(
   ['Date']
).agg(
    {
         'danceability':'mean',
         'energy':'mean',
         'loudness':'mean',
         'cases':'mean',
         'duration_ms':'mean'
    }
)
weekly_stats['duration_min'] = weekly_stats['duration_ms']/60000
weekly_stats['duration_min'].plot(kind='line', 
                figsize=(12,5), 
                title='Song Duration (minutes) throughout 2020',
               colormap='winter');
```

#### What are the one-hit wonders? Could we identify? They were in Top5 only once in the year.

The table below summarizes the so called one-hit wonders. e defined one-hit wonders as songs that were featured in the Top 5 only once. We determined Top 5 as the limit to enter to be a successful song. It is interesting to see that 14 songs cracked the top 5 then never got featured again in the top 5. 

We can also see that most of the top songs are rather danceable with 13 out of the featured 14 songs being above 0.5 for the value of danceability.

```python
temp = df.groupby(['trackID']).size().reset_index().rename(columns = {0:'num_features'}) 
temp = temp[temp['num_features'] == 1].drop('num_features', axis = 1) # filtering for only one occurance songs
oneHit_wonders = df[df['trackID'].isin(temp['trackID'])]
oneHit_wonders[oneHit_wonders['Position'] < 5][['Position','Track Name','Artist','Date', 'danceability']].sort_values('Position',ascending = True)
```
